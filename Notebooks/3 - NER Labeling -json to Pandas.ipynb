```python
import pandas as pd
import json
from google.colab import drive
drive.mount('/content/drive')
```

```python


# Load the JSON data
with open('/content/drive/MyDrive/3owais_Capital/NER_DATA/transcript[2]anno.json', 'r') as json_file:
    data = json.load(json_file)

# Initialize lists to store data
sentences = []
words = []
entities = []

# Iterate through the annotations
for annotation in data['annotations']:
    if annotation:
        sentence = annotation[0].strip()
        entities_data = annotation[1]['entities']
        sentence_words = sentence.split()

        for idx, word in enumerate(sentence_words):
            word_entities = [entity[2] for entity in entities_data if entity[0] <= idx < entity[1]]
            if not word_entities:
                word_entities = ['O']
            sentences.append(sentence)
            words.append(word)
            entities.append(','.join(word_entities))

# Create a DataFrame
ner_data = pd.DataFrame({'Sentence': sentences, 'Words': words, 'Entities': entities})

# Display the DataFrame
len(ner_data)
```
